# Target Analysis

Class imbalance occurs when the number of observations in one class is significantly higher than the other classes. It's a common problem in machine learning and can affect the model accuracy. One common approach of solving this problem is to either decrease the number of samples in the majority class or increase the number of samples in the minority class. Target analysis helps to visualise the class imbalance in the dataset.

```python
import matplotlib
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

from sklearn_evaluation import plot
from sklearn import datasets
```

```python
matplotlib.rcParams["figure.figsize"] = (7, 7)
matplotlib.rcParams["font.size"] = 18
```

```python
# Get the iris dataset and split into train and test sets
iris = load_iris()
X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

##### Balance Mode

When only training data is passed the balance mode is displayed which shows distribution of each class.

```python
plot.target_plot(y_train)
```

##### Balance Mode

When both the training and the test sets are passed, a side by side bar chart of both the sets is displayed.

```python
plot.target_plot(y_train, y_test)
```
